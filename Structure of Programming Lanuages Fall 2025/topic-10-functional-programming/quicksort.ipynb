```python
from pprint import pprint
```

```python
def head(t):
    return t[0]

def tail(t):
    return t[1:]

assert head([1,2,3]) == 1

assert tail([1,2,3]) == [2,3]

assert tail([]) == []
```

```python
T = [6,3,4,2,6,4,5,8,4,2,7]
```

```python
def equal_to(n,t):
    if t == []:
        return []
    if head(t) == n:
        return [ head(t) ] + equal_to(n, tail(t))
    else:
        return equal_to(n, tail(t))

pprint(equal_to(4,T))

assert equal_to(4,T) == [4,4,4]
```

```python
def less_than(n,t):
    if t == []:
        return []
    if head(t) < n:
        return [ head(t) ] + less_than(n, tail(t))
    else:
        return less_than(n, tail(t))

pprint(less_than(4,T))

assert less_than(4,T) == [3,2,2]
```

```python
def greater_than(n,t):
    if t == []:
        return []
    if head(t) > n:
        return [ head(t) ] + greater_than(n, tail(t))
    else:
        return greater_than(n, tail(t))

pprint(greater_than(4,T))

assert greater_than(4,T) == [6, 6, 5, 8, 7]
```

```python
def sort(t):
    if t == []:
        return []
    else:
        return sort(less_than(head(t),t)) + equal_to(head(t),t) + sort(greater_than(head(t),t))

pprint(sort(T))  
```
